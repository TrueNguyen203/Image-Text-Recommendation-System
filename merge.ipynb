```python
import pandas as pd 
```

```python
amazon = pd.read_csv('amazon_clothing.csv')
shein = pd.read_csv('shein_clothing.csv')
walmart = pd.read_csv('walmart_clothing.csv')
```

```python
final_columns = [
    'Product_Name', 'Brand', 'Description', 'Categories', 'Product_ID',
    'Initial_Price', 'Final_Price', 'Currency', 'Rating', 'Reviews_Count', 
    'URL', 'Marketplace'
]
```

```python
amazon = amazon[['title', 'brand', 'description', 'categories', 'asin', 
                'initial_price', 'final_price', 'currency', 'rating', 
                'reviews_count', 'url']].rename(columns={
    'title': 'Product_Name',
    'brand': 'Brand',           
    'description': 'Description', 
    'categories': 'Categories',   
    'asin': 'Product_ID',
    'initial_price': 'Initial_Price', 
    'final_price': 'Final_Price',   
    'currency': 'Currency',         
    'rating': 'Rating',           
    'reviews_count': 'Reviews_Count',
    'url': 'URL'                    
})
amazon['Marketplace'] = 'Amazon'
amazon = amazon[final_columns]

amazon.head(5)
```

```python
shein = shein[['product_name', 'brand', 'description', 'category_tree', 'product_id', 
               'initial_price', 'final_price', 'currency', 'rating', 
               'reviews_count', 'url']].rename(columns={
    'product_name': 'Product_Name',
    'brand': 'Brand',
    'description': 'Description',
    'category_tree': 'Categories', 
    'product_id': 'Product_ID',
    'initial_price': 'Initial_Price',
    'final_price': 'Final_Price',
    'currency': 'Currency',
    'rating': 'Rating',
    'reviews_count': 'Reviews_Count',
    'url': 'URL'
})
shein['Marketplace'] = 'Shein'
shein = shein[final_columns]

shein.head(5)
```

```python
walmart = walmart[['product_name', 'brand', 'description', 'categories', 'product_id', 
                   'initial_price', 'final_price', 'currency', 'rating', 
                   'review_count', 'url']].rename(columns={
    'product_name': 'Product_Name',
    'brand': 'Brand',
    'description': 'Description',
    'categories': 'Categories',
    'product_id': 'Product_ID',
    'initial_price': 'Initial_Price',
    'final_price': 'Final_Price',
    'currency': 'Currency',
    'rating': 'Rating',
    'review_count': 'Reviews_Count', 
    'url': 'URL'
})
walmart['Marketplace'] = 'Walmart'
walmart = walmart[final_columns]

walmart.head(5)
```

```python
merged = pd.concat([amazon, shein, walmart], ignore_index=True)
merged.info()
```
