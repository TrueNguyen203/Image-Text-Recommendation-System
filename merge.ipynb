```python
final_columns = [
    'Product_Name', 'Brand', 'Description', 'Categories', 
    'Initial_Price', 'Final_Price', 'Currency', 'Rating', 'Reviews_Count', 
    'Image_URL'
]
```

```python
amazon = amazon[['title', 'brand', 'description', 'categories', 
                'initial_price', 'final_price', 'currency', 'rating', 
                'reviews_count', 'image_url']].rename(columns={
    'title': 'Product_Name',
    'brand': 'Brand',           
    'description': 'Description', 
    'categories': 'Categories',  
    'initial_price': 'Initial_Price', 
    'final_price': 'Final_Price',   
    'currency': 'Currency',         
    'rating': 'Rating',           
    'reviews_count': 'Reviews_Count',
    'image_url': 'Image_URL'                    
})
amazon = amazon[final_columns]

amazon.head(5)
```

```python
shein = shein[['product_name', 'brand', 'description', 'category_tree',  
               'initial_price', 'final_price', 'currency', 'rating', 
               'reviews_count', 'main_image']].rename(columns={
    'product_name': 'Product_Name',
    'brand': 'Brand',
    'description': 'Description',
    'category_tree': 'Categories', 
    'initial_price': 'Initial_Price',
    'final_price': 'Final_Price',
    'currency': 'Currency',
    'rating': 'Rating',
    'reviews_count': 'Reviews_Count',
    'main_image': 'Image_URL'
})
shein = shein[final_columns]

shein.head(5)
```

```python
walmart = walmart[['product_name', 'brand', 'description', 'categories', 
                   'initial_price', 'final_price', 'currency', 'rating', 
                   'review_count', 'main_image']].rename(columns={
    'product_name': 'Product_Name',
    'brand': 'Brand',
    'description': 'Description',
    'categories': 'Categories',
    'initial_price': 'Initial_Price',
    'final_price': 'Final_Price',
    'currency': 'Currency',
    'rating': 'Rating',
    'review_count': 'Reviews_Count', 
    'main_image': 'Image_URL'
})
walmart = walmart[final_columns]

walmart.head(5)
```
